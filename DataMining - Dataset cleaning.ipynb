```python
import pandas as pd
import numpy as np

header_list =["Age","Workclass","Fnlwgt","Education","Education-num",
              "Marital-status","Occupation","Relationship","Race","Sex",
              "Capital-gain","Capital-loss","Hours-per-week","Native-country","Income"]
```

```python
training_file = pd.read_csv("adult.data.csv", header=None, names=header_list)
training_file = training_file[["Age","Workclass","Fnlwgt","Education","Education-num",
              "Marital-status","Occupation","Relationship","Race","Sex",
              "Capital-gain","Capital-loss","Hours-per-week","Native-country","Income"]].apply(lambda x:x).replace(" ?", np.nan)
print(training_file.head(16))
training_file.to_csv("adult_training_file.csv",index=False)
```
